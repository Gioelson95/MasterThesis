# Features

```python
import librosa
```

### Spectral features

```python
# Spectral centroid, return centroid frequencies

def spectral_centroid(audio):
    cent = librosa.feature.spectral_centroid(y=audio, sr=sr)[0]
    return cent

def spectral_rolloff(audio):
    rolloff = librosa.feature.spectral_rolloff(y=audio, sr=sr)
    return rolloff

def spectral_flatness(audio):
    flatness = librosa.feature.spectral_flatness(y=audio)
    return flatness

def MFCC(audio, n_mfcc):
    mfccs = librosa.feature.mfcc(y=audio, sr=sr, n_mfcc=n_mfcc)
    return mfccs

# Zero Crossing Rate ZCR

def zcr(audio):
    zero_crossings = librosa.zero_crossings(x, pad=False)
    return sum(zero_crossings)

def zcr1(audio):
    count_zero = np.sum(np.abs(np.diff(np.sign(audio)))) / 2
    return count_zero

def zcr2(audio):
    zero_crosses = np.nonzero(np.diff(audio > 0))[0]
    return zero_crosses.size
```

### Rythm features

```python
# tempogram: local autocorrelation of the onset strength envelope
def tempogram(audio, hop_length):
    oenv = librosa.onset.onset_strength(y=audio, sr=sr, hop_length=hop_length)
    tempogram = librosa.feature.tempogram(onset_envelope=oenv, sr=sr,hop_length=hop_length)
    return tempogram

# tempo & beat tracking

def tempo_beat_track(audio, hop_length):
    tempo, beats = librosa.beat.beat_track(y=audio, sr=sr)
    onset_env = librosa.onset.onset_strength(audio, sr=sr, aggregate=np.median)
    tempo, beats = librosa.beat.beat_track(onset_envelope=onset_env, sr=sr)
    
    return tempo, beats
```

```python
# energy

def signal_energy(audio):
    return np.sum(audio**2)
```
