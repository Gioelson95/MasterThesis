# Spectrogram

### This notebook takes the .mp3 files in the directory 'audio_path' and extract the spectrogram and save in a .png file

```python
import librosa, librosa.display
import IPython.display as ipd
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl

from natsort import natsorted
import os
from scipy import signal
```

```python
audio_path = '/Users/gioelepozzi/Desktop/MasterThesis/code/spectrogram/data/1.mp3'

x, sr = librosa.load(audio_path)
ipd.Audio(x, rate=sr)
```

```python
# audio specs

print('Audio Sampling Rate: '+str(sr)+' samples/sec')
print('Total Samples: '+str(np.size(x)))
secs=np.size(x)/sr
print('Audio Length: '+str(secs)+' s')
plt.figure(figsize=(20, 5))
librosa.display.waveplot(x, sr=sr, alpha=0.5)
```

```python
X = librosa.stft(x)
print('This STFT has',X.shape[0],'frequency bins and',X.shape[1],'frames in time.')

# log amplitude
XdB = librosa.amplitude_to_db(abs(X))
```

### spectrogram with librosa (~320 ms)

```python
%%time
plt.figure(figsize=(20, 5))
librosa.display.specshow(XdB, sr=sr, x_axis='time', y_axis='hz')
plt.colorbar(format='%+2.0f dB')
```

```python
plt.figure(figsize=(20, 5))
librosa.display.specshow(XdB, sr=sr)
plt.savefig('1.png', bbox_inches='tight')
```

### spectrogram with matplotlib.pyplot.spectrogram (~150 ms)

```python
%%time
plt.figure(figsize=(20, 5))
plt.specgram(x, NFFT=X.shape[0], Fs=sr)
plt.colorbar(format='%+2.0f dB')
```

# Function to create spectrogram and save in a .png

```python
def spectrogram(audio_path, img_path):
    for mp3_file in natsorted(os.listdir(audio_path)):
        if mp3_file[-4:] == '.mp3':
            file_name = os.path.basename(mp3_file)
            music_ID = file_name.split('.')[0]
            print('filename',file_name,'music_ID',music_ID)
            x, sr = librosa.load(audio_path + '/'+ file_name)
            X = librosa.stft(x)
            plt.specgram(x, NFFT=X.shape[0], Fs=sr)
            plt.savefig(img_path + '/' + music_ID + '.png', bbox_inches='tight') 
```

```python
audio_path = '/Users/gioelepozzi/Desktop/MasterThesis/code/spectrogram/data'
img_path = '/Users/gioelepozzi/Desktop/MasterThesis/code/spectrogram/img'

spectrogram(audio_path, img_path)
```

```python

def spectrogram_audio(file_path, img_path):
    for file in natsorted(os.listdir(file_path)):
        if file[-4:] == '.mp3':
            file_name = os.path.basename(file)
            music_ID = file_name.split('.')[0]
            #print('filename',file_name,'audio','music_ID',music_ID)
            x, sr = librosa.load(file_path + '/'+ file_name)
            X = librosa.stft(x)
            fig, ax = plt.subplots()
            ax.specgram(x, NFFT=X.shape[0], Fs=sr)
            plt.axis('off')
            fig.subplots_adjust(bottom = 0)
            fig.subplots_adjust(top = 1)
            fig.subplots_adjust(right = 1)
            fig.subplots_adjust(left = 0)
            ax.get_xaxis().set_visible(False)
            ax.get_yaxis().set_visible(False)
            fig.savefig(img_path + '/' + music_ID + '_audio' + '.png', bbox_inches='tight', pad_inches=0) 

def spectrogram_eda(file_path, img_path):
    for file in natsorted(os.listdir(file_path)):

        if file[-4:] == '.csv':
            file_name = os.path.basename(file)
            music_ID = file_name.split('_')[0]
            #print('filename',file_name,'eda','music_ID',music_ID)
            eda_df = pd.read_csv(file_path + '/' + file_name)
            ar = eda_df.mean(axis=1)
            ar = np.array(ar)
            fig, ax = plt.subplots() 
            f, t, Sxx = signal.spectrogram(ar)
            ax.pcolormesh(t, f, Sxx, norm=mpl.colors.LogNorm(vmin=Sxx.min(), vmax=Sxx.max()), cmap='inferno')
            plt.axis('off')
            fig.subplots_adjust(bottom = 0)
            fig.subplots_adjust(top = 1)
            fig.subplots_adjust(right = 1)
            fig.subplots_adjust(left = 0)
            fig.savefig(img_path + '/' + music_ID + '_eda' + '.png')
```

```python
file_path = '/Users/gioelepozzi/Desktop/MasterThesis/code/spectrogram/data'
img_path = '/Users/gioelepozzi/Desktop/MasterThesis/code/spectrogram/img'

spectrogram_audio(file_path, img_path)
spectrogram_eda(file_path, img_path)
```
